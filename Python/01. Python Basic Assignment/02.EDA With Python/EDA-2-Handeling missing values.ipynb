## handeling missing data

```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Admin.DESKTOP-Q9CVR13/Desktop/DataSet imarticus/PYTHON/Day 3/weather_data.csv",parse_dates=['day'])
df
```

```python
df.dtypes
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
df_new = df.fillna(0)
df_new
```

```python
df_new = df.fillna({
    'temperature' : 0 ,
    'windspeed' : 0,
    'event' : 'No Event'
})
df_new
```

```python
df_new = df.fillna(method='ffill') #taking value of forward fill
df_new
```

```python
df_new = df.fillna(method='bfill') #taking value of backward fill
df_new 
```

```python
df_new = df.fillna(method='bfill',axis='columns') # column wise fill backward just for checking 
df_new
```

```python
df_new = df.fillna(method='ffill', limit=1) #taking value of forward fill #only just replacing with 1 NAs
df_new
```

```python
# interpolate
df_new = round(df.interpolate()) # rounding and interpolate is taking value as step by step for ed 32,33,34 etc
df_new
```

```python
df_new = df.dropna(how='all') #droping all value which has only all columns NAN
df_new
```

```python
df
```

```python
df_new = df.dropna(thresh=1) # keep those rows having atlest 1 values NON NA, delete others
df_new #jitna threshold likha he utna value hona chaiye
```
